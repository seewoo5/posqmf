```sage
def MLDE1(F):
    w = F.weight()
    DF = F.derivative()
    DDF = DF.derivative()
    DDDF = DDF.derivative()

    DE2 = E2.derivative()
    DE4 = E4.derivative()
    DDE2 = DE2.derivative()
    return DDDF - (w/4) * E2 * DDF + ((w-4)/4 * E4 + w*(w-1)/4 * DE2) * DF - ((w-2)*(w-4)/16 * DE4 + w*(w-1)*(w-2)/24 * DDE2) * F
```

```sage
for w in range(16, 80, 4):
    Fw = pos_eigenform(w)
    print(w, MLDE1(Fw))
```
