## Feigenbaum-Grabner-Hardin

The following shows the first few examples of FGH's modular forms for the $(-1)^{d/4}$ and $(-1)^{d/4 + 1}$-eigenfunctions, which we denote as $F_w$ and $G_w$. Note that $F_w$ is a weight $w$, depth $2$, and level $1$ quasimodular form, while $G_w$ is a form of $\widetilde{G}_{w-12} \Delta \mathcal{L}_S + \Psi_w$, where $\widetilde{G}_w$ (resp. $\Psi_w$) are modular forms of weight $w$ and level 1 (resp. $2$), and $\mathcal{L}_S = \log \lambda_S$. Implementations can be found in `fgh.sage`. Note that they are normalized so that the first nonzero coefficient is 1.

### $(-1)^{d/4}$-family $F_w$

$F_w$ for $8 \le w \le 20$.

```python
for w in range(8, 22, 2):
    Fw = F(w)
    print_qm(Fw, f"F_{w}", w//4 + 10)
```

They satisfy the following recurrence relations: for $w \equiv 0 \pmod{4}$,
$$
\begin{align*}
F_{w+2} &= \frac{(w-8)(w-4)}{768(w-7)(w-5)} (E_4 F_{w - 2} - E_6 F_{w - 4}) \\
\partial_{w - 2} F_{w} &= \frac{w - 5}{6} F_{w + 2}.
\end{align*}
$$

```python
# Sanity check for the recurrence relations
for w in tqdm(range(12, 102, 4)):
    Fwm4 = F(w - 4)
    Fwm2 = F(w - 2)
    Fw = F(w)
    Fwp2 = F(w + 2)

    assert Fwp2 == ((w - 8) * (w - 4)) / (768 * (w - 7) * (w - 5)) * (E4 * Fwm2 - E6 * Fwm4)
    assert qm_serre_derivative(Fw, w - 2) == (w - 5) / 6 * Fwp2
```

$\{F_w\}$ and extremal quasimodular forms $\{X_{w, 2}\}$ are related by the following formula:
$$
\begin{align*}
    F_{w} &=  - \frac{256(w-3)(w-2)(w-1)}{(w-4)w^2} X_{w, 2} + E_4 X_{w-4,2} \\
    F_{w+2} &= \frac{2(w - 3)}{3(w-4)} E_4 X_{w-2, 2} + \frac{w-6}{3(w-4)} E_6 X_{w-4, 2}.
\end{align*}
$$

```python
# Sanity check for the relation between $F_w$ and $X_{w, 2}$
for w in tqdm(range(12, 102, 4)):
    Fw = F(w)
    Fwp2 = F(w + 2)
    Xw2 = extremal_qm(w, 2)
    Xw2m2 = extremal_qm(w - 2, 2)
    Xw2m4 = extremal_qm(w - 4, 2)

    assert Fw ==  - (256 * (w - 3) * (w - 2) * (w - 1)) / ((w - 4) * w^2) * Xw2 + E4 * Xw2m4
    assert Fwp2 == (2 * (w - 3)) / (3 * (w - 4)) * E4 * Xw2m2 + (w - 6) / (3 * (w - 4)) * E6 * Xw2m4
```

Define a differential operator $L_{2, k}$ by
$$
L_{2, k} = \partial_{k}^{2} - \frac{k(k+2)}{144} E_4.
$$
Then $F_w$ and $X_{w,2}$ are related by
$$
\begin{align*}
L_{2, w-2} F_{w} &= \left(\left\lfloor \frac{w}{4} \right\rfloor - 1\right) \left(\left\lfloor \frac{w}{4} \right\rfloor - \frac{w+5}{6}\right) \Delta X_{w-8, 2} \\
L_{2, w-2} X_{w, 2} &= \left\lfloor \frac{w}{4} \right\rfloor \left(\left\lfloor \frac{w}{4} \right\rfloor - \frac{w-1}{6}\right) F_{w+4}
\end{align*}
$$

```python
def qm_L2(k, qm):
    return qm_serre_derivative_fold(qm, 2, k) - k * (k + 2) / 144 * E4 * qm
```

```python
# Sanity check for the relation between $F_w$ and $X_{w, 2}$ with $L_{2,k}$
for w in tqdm(range(12, 102, 4)):
    Fw = F(w)
    Fwp4 = F(w + 4)
    Xw2 = extremal_qm(w, 2)
    Xw2m8 = extremal_qm(w - 8, 2)

    assert qm_L2(w - 2, Fw) == (w // 4 - 1) * (w // 4 - (w + 5) / 6) * Disc * Xw2m8
    assert qm_L2(w - 2, Xw2) == (w // 4) * (w // 4 - ((w - 1) / 6)) * Fwp4
```

Check nonnegativity of the first $n_+$ coefficients of $\widetilde{F}_{w_+ -2} = \frac{\partial F_{w_+}}{\partial E_2}$ for the nonpositivity of $M_{d,+}(\mathbf{0})$. Here $d, w_+, n_+$ are related as
$$
    w_+ = \begin{cases}
        \frac{d}{4} + 16 & d \equiv 0 \pmod{16} \\ \frac{d}{4} + 10 & d \equiv 8 \pmod{16}
    \end{cases}, \quad n_+ = \begin{cases}
        \frac{d}{16} + 1 = \frac{w_+}{4} - 3 & d \equiv 0 \pmod{16} \\ \frac{d-8}{16} + 1 = \frac{w_+}{4} - 2  & d \equiv 8 \pmod{16}
    \end{cases}.
$$
Note that all weights $w_+ \ge 16$ appear twice among all $d \equiv 0 \pmod{8}$, where the corresponding $n_+$ is differ by one.
Hence, it is more efficient to loop over $w_+$ instead of $d$. In this case, the corresponding (bigger) $n_+$ is $w_+ / 4 - 2$ and $d$'s are $4(w_+ - 16)$ and $4(w_+ - 10)$.

```python
# w_max = 9016, which corresponds to d_max = 36000. This would take weeks to finish with current implementation.
w_max = 1016
d_max = 4 * (w_max - 16)
for w in tqdm(range(12, w_max + 1, 4)):
    tFw = Ftilde(w - 2)
    n = w // 4 - 2
    coeffs = qm_coefficients(tFw, n + 1)[1:] # remove the constant term which is 0
    for i, c in enumerate(coeffs, start=1):
        if c < 0:
            raise ValueError(f"Negative coefficient found: w={w}, n={n}, i={i}, c={c}")
print(f"All coefficients are nonnegative. The new upper bound is true up to d={d_max}.")
```

### $(-1)^{d/4+1}$-family $G_w$

$G_w$ for $4 \le w \le 20$.

```python
for w in range(4, 22, 2):
    Gw = G(w)
    print_ls(Gw, f"G_{w}", w//4 + 10)
```

It has a form of
$$
G_w = \widetilde{G}_{w-12} \Delta \mathcal{L}_S + \Psi_w,
$$
where $\widetilde{G}_{w-12}$ (resp. $\Psi_w$) are modular forms of weight $w-12$ (resp. $w$) and level 1 (resp. $\Gamma(2)$), respectively. Note that $\widetilde{G}_{w-12}$ also satisfies similar recurrence relations as $G_w$'s.

```python
# Sanity check for the relation between $G_w$ and $\widetilde{G}_{w-12}$.
for w in tqdm(range(12, 102, 4)):
    Gw = G(w)
    A, B = ls_components(Gw)
    Gtilde_wm12 = Gtilde(w - 12)
    Disc_Gtilde_wm12 = Disc * Gtilde_wm12
    assert l1_to_l2(Disc_Gtilde_wm12) == B
```

They satisfy the following recurrence relations: for $w \equiv 0 \pmod{4}$,
$$
\begin{align*}
    G_{w+2} &= \frac{(w-6)(w-2)}{768(w-5)(w-3)} (E_4 G_{w-2} -E_6 G_{w-4})  \\
    \partial_{w}G_{w} &= \frac{w-3}{6} G_{w + 2}
\end{align*}
$$

```python
# Sanity check for the recurrence relations
for w in tqdm(range(8, 102, 4)):
    Gwm4 = G(w - 4)
    Gwm2 = G(w - 2)
    Gw = G(w)
    Gwp2 = G(w + 2)

    assert Gwp2 == ((w - 6) * (w - 2)) / (768 * (w - 5) * (w - 3)) * (E4_LS * Gwm2 - E6_LS * Gwm4)
    assert ls_serre_derivative(Gw, w) == (w - 3) / 6 * Gwp2
```

$\{Y_{w}\}$ is a companion of $\{X_{w, 2}\}$ which are closely related to $\{G_w\}$.
$Y_w$ for $2 \le w \le 20$.

```python
for w in range(4, 22, 2):
    if w == 6:  # Y_6 = 0
        continue
    Yw = Y(w)
    print_ls(Yw, f"Y_{w}", w//2 + 10)
```

For $w \equiv 0 \pmod{4}$, $Y_w$ satisfies the modular linear differential equation
$$
Y_{w}''' - \frac{w+2}{4} E_2 Y_{w}'' + \frac{(w+1)(w+2)}{4} E_2' Y_{w}' - \frac{w(w+1)(w+2)}{24} E_2'' Y_w = 0.
$$

```python
# Sanity check for the modular linear differential equation satisfied by $Y_w$.
DE2_LS = ls_derivative(E2_LS)
DDE2_LS = ls_derivative(DE2_LS)

for w in tqdm(range(8, 102, 4)):
    Yw = Y(w)
    DYw = ls_derivative(Yw)
    DDYw = ls_derivative(DYw)
    DDDYw = ls_derivative(DDYw)

    assert DDDYw == (w + 2) / 4 * E2_LS * DDYw - (w + 1) * (w + 2) / 4 * DE2_LS * DYw + w * (w + 1) * (w + 2) / 24 * DDE2_LS * Yw
```

$G_w$ and $Y_w$ are related by
$$
\begin{align*}
    Y_{w+2} &= \frac{3(w+2)^{2}}{16(w-4)^{2}(w+1)(w+3)} \left(\frac{(w-3)(w-2)}{36} E_4 Y_{w-2} - \partial_{w-2}^{2} Y_{w-2}\right) \\
    Y_{w+2} &= \frac{(w+2)^2}{768(w+1)(w+3)} (E_{4} Y_{w-2} - E_{6} Y_{w-4}) \\
    G_{w} &= -\frac{256(w-1)w(w+1)}{(w-2)(w+2)^{2}} Y_{w} + E_{4} Y_{w-4} \\
    G_{w+2} &= \frac{2(w-1)}{3(w-2)} E_{4} Y_{w-2} + \frac{w-4}{3(w-2)} E_{6} Y_{w-4}
\end{align*}
$$

```python
# Sanity check for the relation between $G_w$ and $Y_w$.
for w in tqdm(range(12, 102, 4)):
    Ywm4 = Y(w - 4)
    Ywm2 = Y(w - 2)
    Yw = Y(w)
    Ywp2 = Y(w + 2)
    SYwm2 = ls_serre_derivative(Ywm2, w - 2)
    SSYwm2 = ls_serre_derivative(SYwm2, w)

    Gw = G(w)
    Gwp2 = G(w + 2)

    assert Ywp2 == (3 * (w + 2)^2) / (16 * (w - 4)^2 * (w + 1) * (w + 3)) * ((w - 3) * (w - 2) / 36 * E4_LS * Ywm2 - SSYwm2)
    assert Ywp2 == ((w + 2)^2) / (768 * (w + 1) * (w + 3)) * (E4_LS * Ywm2 - E6_LS * Ywm4)
    assert Gw == -(256 * (w - 1) * w * (w + 1)) / ((w - 2) * (w + 2)^2) * Yw + E4_LS * Ywm4
    assert Gwp2 == (2 * (w - 1)) / (3 * (w - 2)) * E4_LS * Ywm2 + (w - 4) / (3 * (w - 2)) * E6_LS * Ywm4
```

The operator $L_{2, k}$ also relates $G_w$ and $Y_w$ by
$$
\begin{align*}
    L_{2, w} G_{w} &= \left(\left\lfloor \frac{w}{4} \right\rfloor - \frac{1}{2}\right) \left(\left\lfloor \frac{w}{4} \right\rfloor - \frac{w + 4}{6}\right) \Delta Y_{w - 8} \\
    L_{2, w} Y_{w} &= \left(\left\lfloor \frac{w}{4} \right\rfloor + \frac{1}{2}\right) \left(\left\lfloor \frac{w}{4} \right\rfloor - \frac{w - 2}{6}\right) G_{w + 4} 
\end{align*}
$$

```python
def ls_L2(k, f):
    Sf = ls_serre_derivative(f, k)
    SSf = ls_serre_derivative(Sf, k + 2)
    return SSf - k * (k + 2) / 144 * E4_LS * f
```

```python
# Sanity check for the relation between $G_w$ and $Y_w$ with $L_{2, k}$.
for w in tqdm(range(12, 102, 4)):
    Ywm8 = Y(w - 8)
    Yw = Y(w)
    Gw = G(w)
    Gwp4 = G(w + 4)

    assert ls_L2(w, Gw) == (w // 4 - 1/2) * (w // 4 - (w + 4) / 6) * Disc_LS * Ywm8
    assert ls_L2(w, Yw) == (w // 4 + 1/2) * (w // 4 - (w - 2) / 6) * Gwp4
```

Check nonnegativity of the first $n_- - 1$ coefficients of $\widetilde{G}_{w_{-} - 12}$ for the nonpositivity of $M_{d,-}(\mathbf{0})$. Here $d, w_-, n_-$ are related as
$$
    w_- = \begin{cases}
        \frac{d}{4} + 11 & d \equiv 4 \pmod{16} \\ \frac{d}{4} + 5 & d \equiv 12 \pmod{16}
    \end{cases},\quad
    n_- = \begin{cases}
        \frac{d + 12}{16} = \frac{w_-}{4} - 2 & d \equiv 4 \pmod{16} \\ \frac{d + 4}{16} = \frac{w_-}{4} - 1 & d \equiv 12 \pmod{16}
    \end{cases}.
$$
Note that all weights $w_- \ge 12$ appear twice among all $d \equiv 4 \pmod{8}$, where the corresponding $n_-$ is differ by one.
Hence, it is more efficient to loop over $w_-$ instead of $d$. In this case, the corresponding (bigger) $n_-$ is $w_- / 4 - 1$ and $d$'s are $4(w_- - 11)$ and $4(w_- - 5)$.

```python
# w_max = 9011, which corresponds to d_max = 36000. This would take weeks to finish with current implementation.
w_max = 1011
d_max = 4 * (w_max - 11)
for w in tqdm(range(12, w_max + 1, 4)):
    tGwm12 = Gtilde(w - 12)
    n = w // 4 - 1
    coeffs = qm_coefficients(tGwm12, n)
    for i, c in enumerate(coeffs, start=1):
        if c < 0:
            raise ValueError(f"Negative coefficient found: w={w}, n={n}, i={i}, c={c}")
print(f"All coefficients are nonnegative. The new upper bound is true up to d={d_max}.")
```

## Extremal Eisenstein Series

Extremal Eisenstein series (or extremal modular form, extremal theta series, depending on literature) is a modular form of a given weight with $q$-expansion $1, 0, ..., 0$ with as many zeros as possible.
We check certain identities of extremal Eisenstein series with small weights, and also check complete positivity of them when weight is a multiple of 4, using Jenkins and Rouse's bound.

```python
for w in range(12, 22, 2):
    XE_w = QM(extremal_eisenstein_series(w))
    print_qm(XE_w, f"XE_{w}")
```

## (Quasi)modular form identities

Identities relating (derivatives of) extremal Eisenstein series and other quasimodular forms, showing that $q$-expansion of them have the same signs (except for the constant term).
This show that the Fourier coefficients of extremal Eisenstein series of weight $\le 20$ have the desired signs (positive for $w \equiv 0 \pmod{4}$, negative for $w \equiv 2 \pmod{4}$).

```python
E_8 = eisenstein(8)
E_10 = eisenstein(10)
E_12 = eisenstein(12)
E_14 = eisenstein(14)

X_14_1 = extremal_qm(14, 1)
X_16_1 = extremal_qm(16, 1)
X_18_1 = extremal_qm(18, 1)
X_20_1 = extremal_qm(20, 1)
X_22_1 = extremal_qm(22, 1)

for w in range(12, 22, 2):
    XE_w = QM(extremal_eisenstein_series(w))
    DXE_w = XE_w.derivative()
    print_qm(XE_w, f"XE_{w}")
    print_qm(DXE_w, f"DXE_{w}")

XE_4 = QM(extremal_eisenstein_series(4))
XE_6 = QM(extremal_eisenstein_series(6))
XE_8 = QM(extremal_eisenstein_series(8))
XE_10 = QM(extremal_eisenstein_series(10))
XE_12 = QM(extremal_eisenstein_series(12))
DXE_12 = XE_12.derivative()
XE_14 = QM(extremal_eisenstein_series(14))
XE_16 = QM(extremal_eisenstein_series(16))
DXE_16 = XE_16.derivative()
XE_18 = QM(extremal_eisenstein_series(18))
DXE_18 = XE_18.derivative()
XE_20 = QM(extremal_eisenstein_series(20))
DXE_20 = XE_20.derivative()

print(qm_find_lin_comb(DXE_12, [X_14_1]))
print(qm_find_lin_comb(DXE_16, [E4 * DXE_12, X_18_1]))
print(qm_find_lin_comb(DXE_18, [X_20_1, E4 * X_16_1]))
print(qm_find_lin_comb(DXE_20, [E4 * DXE_16, X_22_1]))

assert XE_4 == E4
assert XE_6 == E6
assert XE_8 == E_8
assert XE_10 == E_10
assert DXE_12 == 393120 * X_14_1
assert XE_14 == E_14
assert DXE_16 == 68/91 * E4 * DXE_12 + 86169600 * X_18_1
assert DXE_18 == - 161336448 * X_20_1 - 172368 * E4 * X_16_1
assert DXE_20 == 55/204 * E4 * DXE_16 + 192192000 * X_22_1
```
