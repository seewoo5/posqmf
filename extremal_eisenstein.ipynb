# Extremal Eisenstein Series

Extremal Eisenstein series (or extremal modular form, extremal theta series, depending on literature) is a modular form of a given weight with $q$-expansion $1, 0, ..., 0$ with as many zeros as possible.
We check certain identities of extremal Eisenstein series with small weights, and also check complete positivity of them when weight is a multiple of 4, using Jenkins and Rouse's bound.

```python
from posqmf import *
from tqdm import tqdm
```

```python
for w in range(12, 22, 2):
    XE_w = QM(extremal_eisenstein_series(w))
    print_qm(XE_w, f"XE_{w}")
```

## (Quasi)modular form identities

Identities relating (derivatives of) extremal Eisenstein series and other quasimodular forms, showing that $q$-expansion of them have the same signs (except for the constant term).
This show that the Fourier coefficients of extremal Eisenstein series of weight $\le 20$ have the desired signs (positive for $w \equiv 0 \pmod{4}$, negative for $w \equiv 2 \pmod{4}$).

```python
E_8 = eisenstein(8)
E_10 = eisenstein(10)
E_12 = eisenstein(12)
E_14 = eisenstein(14)

X_14_1 = extremal_qm(14, 1)
X_16_1 = extremal_qm(16, 1)
X_18_1 = extremal_qm(18, 1)
X_20_1 = extremal_qm(20, 1)
X_22_1 = extremal_qm(22, 1)

for w in range(12, 22, 2):
    XE_w = QM(extremal_eisenstein_series(w))
    DXE_w = XE_w.derivative()
    print_qm(XE_w, f"XE_{w}")
    print_qm(DXE_w, f"DXE_{w}")

XE_4 = QM(extremal_eisenstein_series(4))
XE_6 = QM(extremal_eisenstein_series(6))
XE_8 = QM(extremal_eisenstein_series(8))
XE_10 = QM(extremal_eisenstein_series(10))
XE_12 = QM(extremal_eisenstein_series(12))
DXE_12 = XE_12.derivative()
XE_14 = QM(extremal_eisenstein_series(14))
XE_16 = QM(extremal_eisenstein_series(16))
DXE_16 = XE_16.derivative()
XE_18 = QM(extremal_eisenstein_series(18))
DXE_18 = XE_18.derivative()
XE_20 = QM(extremal_eisenstein_series(20))
DXE_20 = XE_20.derivative()

print(qm_find_lin_comb(DXE_12, [X_14_1]))
print(qm_find_lin_comb(DXE_16, [E4 * DXE_12, X_18_1]))
print(qm_find_lin_comb(DXE_18, [X_20_1, E4 * X_16_1]))
print(qm_find_lin_comb(DXE_20, [E4 * DXE_16, X_22_1]))

assert XE_4 == E4
assert XE_6 == E6
assert XE_8 == E_8
assert XE_10 == E_10
assert DXE_12 == 393120 * X_14_1
assert XE_14 == E_14
assert DXE_16 == 68/91 * E4 * DXE_12 + 86169600 * X_18_1
assert DXE_18 == - 161336448 * X_20_1 - 172368 * E4 * X_16_1
assert DXE_20 == 55/204 * E4 * DXE_16 + 192192000 * X_22_1
```
