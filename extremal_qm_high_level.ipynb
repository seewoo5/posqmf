# Extremal quasimodular forms of higher levels

This notebook contains computations on extremal quasimodular forms of level $\Gamma_0(N)$ for $N = 2, 3, 4$ and depth $\le 1$ (and also depth 2 for $N = 2$).

```python
from posqmf import *
from tqdm import tqdm
```

## Uniqueness

The following code checks the uniqueness of extremal quasimodular forms of level $\Gamma_0(N)$ for $N = 2, 3$, depth $\le 4$, and weight $w \le 100$. `is_extremal_qm_unique_level` computes the coefficient matrix of the basis of the corresponding space of quasimodular forms and checks if it is invertible.

```python
w_max = 50
s_max = 4
for N in [2, 3]:
    for s in range(1, s_max + 1):
        for w in tqdm(range(2, w_max + 1, 2)):
            assert is_extremal_qm_unique_level(w, s, level=N)
        print(f"All extremal quasimodular forms of weight {w} \le {w_max} and depth {s} for Gamma0({N}) are unique.")
```

### Depth 1

Extremal quasimodular forms $\mathcal{D}_{w,1}$ of weight $w$, depth $1$, and level $2$ for $w \le 20$. They have vanishing order $w/2$ at the cusp.

```python
for w in range(2, 22, 2):
    cD_w = extremal_qm_l2(w, 1)
    print_qm(cD_w, f"D_{w},1")
```

$\mathcal{D}_{w,1}$ satisfy the following modular linear differential equation:
$$
\mathcal{D}_{w,1}'' - \frac{w}{2} E_{2}^{[2]} \mathcal{D}_{w,1}' + \frac{w(w-1)}{4} (E_{2}^{[2]})' \mathcal{D}_{w,1} = 0.
$$
From this, the Fourier expansion of $\mathcal{D}_{w,1}$ can be written as
$$
\mathcal{D}_{w,1} = q^{\frac{w}{2}} + \sum_{k \ge 1} f_k(w) q^{\frac{w}{2} + k},
$$
where for each $k \ge 1$, $f_k(w)$ is a rational function in $w$, satisfying the recurrence relation
$$
f_{k}(w) = \frac{1}{\left(\frac{w}{2} + k\right)k} \left( \sum_{i=1}^{k} \left(\frac{w}{2} \left(\frac{w}{2} + (k - i)\right) - \frac{w(w-1)i}{4}\right) c_{i} f_{k-i}(w)\right)
$$

```python
# Sanity check for modular linear differential equation
for w in tqdm(range(2, 52, 2)):
    cD_w = extremal_qm_l2(w, 1)
    DcD_w = cD_w.derivative()
    DDcD_w = DcD_w.derivative()
    DE2_u2_l2 = E2_u2_l2.derivative()
    assert DDcD_w == (w/2) * E2_u2_l2 * DcD_w - w * (w - 1) / 4 * DE2_u2_l2 * cD_w
```

```python
K.<t> = QQ[]

def b(i):
    if i == 0:
        return 1
    else:
        if i % 2 == 0:
            return 8 * sigma(i, 1) - 32 * sigma(i // 2, 1)
        else:
            return 8 * sigma(i, 1)

def f(k):
    if k == 0:
        return K(1)
    else:
        r = K(0)
        for i in range(1, k + 1):
            r += (t / 2 * (t / 2 + (k - i)) - t * (t-1) * i / 4) * b(i) * f(k - i)
        r *= 1 / (t / 2 + k) / k
        return r

for k in range(10):
    print(k, f(k).factor())

# Sanity check for the formula of $f_k(w)$
M = 10
for w in tqdm(range(2, 42, 2)):
    cD_w = extremal_qm_l2(w, 1)
    coeffs = cD_w.coefficients(range(w // 2, w // 2 + M))
    assert coeffs == [f(k)(w) for k in range(M)]
```

They satisfy the following recurrence relations:

$$
\begin{align*}
    \mathcal{D}_{w+4,1} &= \frac{(w+2)(w+4)}{16(w+3)^2} (A_2 \mathcal{D}_{w+2,1} - A_{4,1} \mathcal{D}_{w,1}) \\
    \mathcal{D}_{w+2,1} &= \frac{w+2}{8(w+1)^2} \left(\frac{5w+1}{12} A_2 \mathcal{D}_{w,1} - \partial_{w-1} \mathcal{D}_{w,1}\right) \\
    \mathcal{D}_{w+2,1}' &= -4(w+1) \mathcal{D}_{2,1}\mathcal{D}_{w+2,1} + \frac{w+2}{2} A_{4,1}\mathcal{D}_{w,1}
\end{align*}
$$

```python
# Sanity check for the recurrence relations
cD_2 = extremal_qm_l2(2, 1)
for w in range(2, 22, 2):
    cD_w = extremal_qm_l2(w, 1)
    cD_wp2 = extremal_qm_l2(w + 2, 1)
    cD_wp4 = extremal_qm_l2(w + 4, 1)

    ScD_w = cD_w.derivative() - (w-1) / 12 * E2_l2 * cD_w
    assert cD_wp4 == ((w + 2) * (w + 4)) / (16 * (w + 3) ** 2) * (A2_l2 * cD_wp2 - A4_1_l2 * cD_w)
    assert cD_wp2 == (w + 2) / (8 * (w + 1) ** 2) * ((5 * w + 1) / 12 * A2_l2 * cD_w - ScD_w)
    assert cD_wp2.derivative() == -4 * (w + 1) * cD_2 * cD_wp2 + (w + 2) / 2 * A4_1_l2 * cD_w
```

$\mathcal{D}_{w, 1}$ has integral Fourier coefficients if and only if $w = 2$.

```python
# Sanity check for modular linear differential equation
for w in tqdm(range(4, 42, 4)):
    cD_w = extremal_qm_l2(w, 2)
    DcD_w = cD_w.derivative()
    DDcD_w = DcD_w.derivative()
    DDDcD_w = DDcD_w.derivative()

    DE2_u2_l2 = E2_u2_l2.derivative()
    DDE2_u2_l2 = DE2_u2_l2.derivative()
    
    assert DDDcD_w == (3*w/4) * E2_u2_l2 * DDcD_w - (3*(w-1)*w/4) * DE2_u2_l2 * DcD_w + ((w-2)*(w-1)*w/8) * DDE2_u2_l2 * cD_w
```

For even $w \ge 2$, there exists a unique (normalized) extremal quasimodular form $\mathcal{E}_{w}$ of weight $w$, depth $1$, and level $\Gamma_0(3)$, which has vanishing order $\lfloor 2w/3 \rfloor$ at the cusp.

For $w \equiv 0 \pmod{6}$, they satisfy the following modular linear differential equation:
$$
\mathcal{E}_{w}'' - \frac{2w}{3} E_2^{[3]} \mathcal{E}_{w}' + \frac{w(w-1)}{3} (E_2^{[3]})' \mathcal{E}_{w} = 0,
$$
From this, the Fourier expansion of $\mathcal{E}_{w}$ can be written as
$$
\mathcal{E}_{w} = q^{\frac{2w}{3}} + \sum_{k \ge 1} g_k(w) q^{\frac{2w}{3} + k},
$$
where for each $k \ge 1$, $g_k(w)$ is a rational function in $w$, satisfying the recurrence relation
$$
g_{k}(w) = \frac{1}{\left(\frac{2w}{3} + k\right)k} \left( \sum_{i=1}^{k} \left(\frac{2w}{3} \left(\frac{2w}{3} + (k - i)\right) - \frac{w(w-1)i}{3}\right) c_{i} g_{k-i}(w)\right)
$$

```python
# Sanity check for the modular linear differential equation for extremal quasimodular forms of level Gamma0(3) and weight divisible by 6
for w in tqdm(range(6, 62, 6)):
    cE_w = extremal_qm_l3(w, 1)
    DcE_w = cE_w.derivative()
    DDcE_w = DcE_w.derivative()
    DE2_u3_l3 = E2_u3_l3.derivative()
    assert DDcE_w == (2*w/3) * E2_u3_l3 * DcE_w - w * (w - 1) / 3 * DE2_u3_l3 * cE_w
```

```python
K.<t> = QQ[]

def c(i):
    if i == 0:
        return 1
    else:
        if i % 3 == 0:
            return 3 * sigma(i, 1) - 27 * sigma(i // 3, 1)
        else:
            return 3 * sigma(i, 1)

def g(k):
    if k == 0:
        return K(1)
    else:
        r = K(0)
        for i in range(1, k + 1):
            r += (2 * t / 3 * (2 * t / 3 + (k - i)) - t * (t-1) * i / 3) * c(i) * g(k - i)
        r *= 1 / (2 * t / 3 + k) / k
        return r

for k in range(10):
    print(k, g(k).factor())

# Sanity check for the formula of $g_k(w)$
M = 10
for w in tqdm(range(6, 62, 6)):
    cE_w = extremal_qm_l3(w, 1)
    coeffs = cE_w.coefficients(range(2*w//3, 2*w//3 + M))
    assert coeffs == [g(k)(w) for k in range(M)]
```

    &\quad\left. + \frac{2(w+3)(w^2 + 6w + 7)}{3(w+4)^2} B_{4,1} \mathcal{E}_{w+2}\right) \\

$$

```python
# Sanity check for the recurrence relations
cE_2 = extremal_qm_l3(2, 1)
for w in tqdm(range(6, 62, 6)):
    cE_w = extremal_qm_l3(w, 1)
    cE_wp2 = extremal_qm_l3(w + 2, 1)
    cE_wp4 = extremal_qm_l3(w + 4, 1)
    cE_wp6 = extremal_qm_l3(w + 6, 1)

    ScE_w = qm_l3_serre_derivative(cE_w, w - 1)
    ScE_wp2 = qm_l3_serre_derivative(cE_wp2, w + 1)
    ScE_wp4 = qm_l3_serre_derivative(cE_wp4, w + 3)

    DcE_wp2 = cE_wp2.derivative()
    DcE_wp4 = cE_wp4.derivative()
    DcE_wp6 = cE_wp6.derivative()

    assert cE_wp2 == (2*w+3) / 9 / (w+1)^2 * (-ScE_w + (7*w+1)/12 * B2_l3 * cE_w)
    assert cE_wp4 == 2 * (w+3) / 9 / (w+2)^2 * (-ScE_wp2 + (7*w+11)/12 * B2_l3 * cE_wp2)
    assert cE_wp6 == 2 * (w+6) * (2*w+9) / 243 / (w+5)^2 * (-ScE_wp4 - (w+3)*(w^2 - 8*w - 56) / 12 / (w+4)^2 * B2_l3 * cE_wp4 + 2 * (w+3) * (w^2 + 6*w + 7) / 3 / (w+4)^2 * B4_1_l3 * cE_wp2)

    assert DcE_wp2 == -3 * (w+1) * cE_2 * cE_wp2 + (2*w/3 + 1) * B4_1_l3 * cE_w
    assert DcE_wp4 == -3 * (w+3) * cE_2 * cE_wp4 + 2*(w+3)/3/(w+2)^2 * ((w^2 + 6*w + 7) * B4_1_l3 * cE_wp2 - (2*w+3) * B6_2_l3 * cE_w)
    assert DcE_wp6 == -3 * (w+5) * cE_2 * cE_wp6 + 4*(w+3)*(w+6)*(2*w+9)/81/(w+4)^2 * (B4_1_l3 * cE_wp4 - B6_2_l3 * cE_wp2)
```

### Victor-Miller basis for $\Gamma_0(3)$

Using extremal modular forms (depth 0) of level $\Gamma_0(3)$, we can construct Victor-Miller basis for modular forms of level $\Gamma_0(3)$, which is a basis $f_0, \dots, f_{d-1}$ such that the $q$-expansion of $f_i = \sum_{n \ge 0} a_n(f_i) q^{n}$ satisfies $a_j(f_i) = \delta_{i, j}$ for $j = 0, \dots, d - 1$.

```python
for w in range(2, 22, 2):
    vm_basis = l3_victor_miller_basis(w)
    print(f"Victor-Miller basis for weight {w} and level Gamma0(3):")
    for i, f in enumerate(vm_basis):
        print(f"f_{i} = {f.q_expansion(10)}")
```
