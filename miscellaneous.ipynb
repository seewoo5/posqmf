### Uniqueness of extremal quasimodular forms of higher depths

The below code checks the uniqueness of extremal quasimodular forms of depth $5 \le s \le 10$ and weight $w \le 200$.

```python
for s in range(5, 11):
    for w in tqdm(range(2 * s, 202, 2)):
        if w == 2 * s + 2:
            continue
        if not is_extremal_qm_unique(w, s):
            print(f"Extremal quasimodular form of weight {w} and depth {s} is NOT unique.")
```

### Complete positivity of extremal quasimodular forms of higher depths

The below code find possible counterexamples to the naive extention of Kaneko--Koike conjecture for depth $s \ge 5$. For each $X_{w, s}$ with cusp order $m = m_{w, s}$, we only check coefficients up to $q^{2m}$.
Hence it is possible that the "passed" cases may have negative coefficients at higher powers of $q$.

```python
def check_all_positive(ls):
    # check if all the elements of a given list is positive
    # if there is a negative element, return it with index
    for i, x in enumerate(ls):
        if x < 0:
            return False, i, x
    return True, None, None 

for s in range(5, 11):
    for w in range(2 * s, 102, 2):
        if w == 2 * s + 2:
            continue
        X_w_s = extremal_qm(w, s)
        N = 2 * qm_cusp_order(X_w_s, 1000)
        check, i, x = check_all_positive(qm_coefficients(X_w_s, prec=N))
        if not check:
            print(f"X_{w},{s} is not completely positive: the {i}-th coefficient is {x} < 0.")

```

$X_{16, 5}$ is the smallest counterexample, where all the coefficients except for the leading term are negative. This is proved by expressing it as a linear combination of derivatives of Eisenstein series and $\Delta E_4$ and $\Delta X_{4, 2}$, whose $q$-expansions can be easily bounded using the bounds for the divisor function $\sigma_k(n)$ and Ramanujan's tau function $\tau(n)$.

```python
X_16_5 = extremal_qm(16, 5)
print_qm(X_16_5, "X_16,5", prec=100)

E8 = QM(eisenstein(8))
E10 = QM(eisenstein(10))
E12 = QM(eisenstein(12))
E14 = QM(eisenstein(14))
E16 = QM(eisenstein(16))

X_4_2 = extremal_qm(4, 2)
DE14 = E14.derivative()
DDE12 = E12.derivative().derivative()
DDDE10 = E10.derivative().derivative().derivative()
DDDDE8 = E8.derivative().derivative().derivative().derivative()
DDDDDE6 = E6.derivative().derivative().derivative().derivative().derivative()

ls = [
    DE14,
    DDE12,
    DDDE10,
    DDDDE8,
    DDDDDE6,
    Disc * E4,
    Disc * X_4_2,
]

cs = qm_find_lin_comb(X_16_5, ls, prec=100)
print(cs)
```

Combined with the trivial bounds $n^k < \sigma_k(n) < 2 n^k$ for $k \ge 2$ and Deligne's bound $|\tau(n)| \le \sigma_0(n) n^{11/2}$, we can prove that the $n$-th coefficient of $X_{16, 5}$ is negative for $n \ge 250$, whose proof is also formalized in `lean/X_16_5_coeff.lean`. We can check for $8 \le n < 250$ separately.

```python
coeffs = X_16_5.coefficients(list(range(8, 250)))
for i, c in enumerate(coeffs, start=8):
    assert c <= 0, f"{i}-th coefficient is positive: {c}"
```
