### Uniqueness of extremal quasimodular forms of higher depths

The below code checks the uniqueness of extremal quasimodular forms of depth $5 \le s \le 10$ and weight $w \le 200$.

```python
for s in range(5, 11):
    for w in tqdm(range(2 * s, 202, 2)):
        if w == 2 * s + 2:
            continue
        if not is_extremal_qm_unique(w, s):
            print(f"Extremal quasimodular form of weight {w} and depth {s} is NOT unique.")
```

### Complete positivity of extremal quasimodular forms of higher depths

The below code find possible counterexamples to the naive extention of Kaneko--Koike conjecture for depth $s \ge 5$.

```python
def check_all_positive(ls):
    # check if all the elements of a given list is positive
    # if there is a negative element, return it with index
    for i, x in enumerate(ls):
        if x < 0:
            return False, i, x
    return True, None, None 

for s in range(5, 11):
    for w in range(2 * s, 202, 2):
        if w == 2 * s + 2:
            continue
        X_w_s = extremal_qm(w, s)
        N = 2 * qm_cusp_order(X_w_s, 1000)
        check, i, x = check_all_positive(qm_coefficients(X_w_s, prec=N))
        if not check:
            print(f"X_{w},{s} is not completely positive: the {i}-th coefficient is {x} < 0.")

```
